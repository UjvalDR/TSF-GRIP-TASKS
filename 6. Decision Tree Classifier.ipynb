# Task 6. Decision Tree Classifier


- Create a Decision Tree classifier and visualize it graphically.
- The purpose is if we feed any new data to this classifier, it would be able to predict the right class accordingly.


### Import the Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets as datasets

plt.rcdefaults()
plt.rcParams.update({'axes.facecolor':'white'})
```

### Create the dataframe

```python
## load the iris dataset
iris=datasets.load_iris()
```

```python
df= pd.DataFrame(iris.data, columns= iris.feature_names)
df
```

```python
df.info()
```

There are no null-values.

```python
## Visualizing the feature variables.
sns.pairplot(df)
```

### Train- Test split

```python
## Getting the feature and target variables.
X=iris.data
Y=iris.target
```

```python
from sklearn.model_selection import train_test_split
X_train, x_test, Y_train, y_test= train_test_split(X, Y,test_size=0.26,random_state=0)
```

### Decision Tree Classification

```python
## import the classifier
from sklearn.tree import DecisionTreeClassifier
```

```python
## fitting the model
dt=DecisionTreeClassifier()
dt.fit(X_train, Y_train)
```

```python
## predicting values
y_pred= dt.predict(x_test)
```

```python
## checking the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_pred,y_test))
```

```python
y_pred
```

### Visualizing the Decision Tree

```python
## Textual Representation
from sklearn import tree
txt_rep=tree.export_text(dt)
txt_rep
```

```python
## tree plot representation
fig = plt.figure(figsize=(25,20))
treeplot = tree.plot_tree(dt, 
                   feature_names=iris.feature_names,  
                   class_names=iris.target_names,
                   filled=True)
```

### Testing with our own values

```python
y=dt.predict([[5.1,3.1,4.95,0.8]])
if(y==0):
    print('Iris- setosa')
elif(y==1):
    print('Iris- versicolor')
else:
    print('Iris- virginica')
```

### Conclusion

The Decision Tree model has 97.4 % accuracy. Feeding entries to the model will predict the iris flower species accordingly.
